```python
import pandas as pd
from collections import Counter

#fonksiyon tanımlama ve yetkinlik kümesi oluşturma
def yetkinlikSeti(veri, meslek):
    yetkinlik = []

    for satır in veri:
        if satır[0] == meslek:
            yetkinlik.extend(satır[2].split(','))
    
    yetkinlik_sayacı = Counter(yetkinlik)
    
    return yetkinlik_sayacı

veri = pd.read_csv('hamVeri.csv')

# Seçilen mesleğin adı
meslekAdı = 'Mobile Application Developer'

# Mesleğin tüm ülkelerdeki yetkinliklerini bir sayacı olarak oluşturma
yetkinlik_sayacı = yetkinlikSeti(veri.values.tolist(), meslekAdı)

print("Seçilen Meslek:", meslekAdı)
print("Yetkinliklerin Tekrar Sayısı:")
for yetkinlik, sayı in yetkinlik_sayacı.items():
    print(yetkinlik, "=", sayı)
```

```python
from collections import Counter

# fonksiyon tanımlama ve yetkinlik kümesi oluşturma
def yetkinlikSeti(veri):
    yetkinlik = []

    for satır in veri:
        if isinstance(satır[2], str):  # sadece string değerleri dikkate al
            yetkinlik.extend(satır[2].split(','))

    yetkinlik_sayacı = Counter(yetkinlik)

    return yetkinlik_sayacı

veri = pd.read_csv('hamVeri.csv')

# Tüm mesleklerdeki yetkinlikleri bir sayacı olarak oluşturma
yetkinlik_sayacı = yetkinlikSeti(veri.values.tolist())

print("Tüm Mesleklerdeki Yetkinliklerin Tekrar Sayısı:")
for yetkinlik, sayı in yetkinlik_sayacı.items():
    print(yetkinlik, "=", sayı)
```

```python
import pandas as pd

def yetkinlikSeti(veri, yetkinlik):
    indeksler = [] #boş küme

    for indeks, satır in veri.iterrows():
        if isinstance(satır['Yetkinlik'], str) and yetkinlik in satır['Yetkinlik']:
            indeksler.append(indeks) 

    return indeksler

veri = pd.read_csv('hamVeri.csv')

yetkinlik = '2'  # Kontrol etmek istenilen yetkinlik

indeksler = yetkinlikSeti(veri, yetkinlik)

if len(indeksler) > 0:
    print(f"'{yetkinlik}' yetkinliği şu indekslerde bulunuyor:")
    for indeks in indeksler:
        print(indeks) 
else:
    print(f"'{yetkinlik}' yetkinliği veri setinde bulunmuyor.")
```

```python
def meslek_mbd(veri):
    deneyim = {} #boş sözlük

    for satır in veri:
        meslekAdı = satır[0]
        ülke = satır[1]
        deneyimYılı = satır[3]

        if pd.notnull(deneyimYılı):  # NaN değeri kontrol et
            deneyimYılı = int(deneyimYılı)
            
            if meslekAdı not in deneyim: #meslekadını deneyim sözlüğünde kontrol eder, 
                deneyim[meslekAdı] = [] #yoksa liste oluşturur.

            deneyim[meslekAdı].append(deneyimYılı)

    return deneyim

# veri setini okuma
veri = pd.read_csv('hamVeri.csv')

# Seçilen meslek
meslekAdı = 'Artificial Intelligence Engineering'

# Seçilen mesleğe ait deneyim yıllarını getirme
deneyimYılları = meslek_mbd(veri.values.tolist())[meslekAdı]

# Sonuc
print("Seçilen Meslek:", meslekAdı)
print("Deneyim Yılları:", deneyimYılları)
```

```python
from collections import Counter

def meslek_mbd(veri, meslekAdı):
    deneyimYılları = []

    for satir in veri:
        if satir[0] == meslekAdı:
            yil = int(satir[3]) if satir[3] != '' else 0
            deneyimYılları.append(yil)

    return deneyimYılları

# veri setini okuma
veri = pd.read_csv('hamVeri.csv')

# Seçilen meslek
meslekAdı = 'Psychiatrist'

# Seçilen mesleğe ait deneyim yıllarını getir
deneyimYılları = meslek_mbd(veri.values.tolist(), meslekAdı)

# En çok tekrar eden deneyim yılını bul
sayac = Counter(deneyimYılları)
tekrarlıYıl = sayac.most_common(1)[0][0] #most_common en çok eden veriyi döndürür

# Sonucu yazdır
print("Seçilen Meslek:", meslekAdı)
print("En Çok Tekrar Eden Deneyim Yılı:", tekrarlıYıl)
```
